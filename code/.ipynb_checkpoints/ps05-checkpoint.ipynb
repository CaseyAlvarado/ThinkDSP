#### 1) Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes.  The usual version of IDFT uses frequencies 0 through N-1.  What happens if you go from N to 2N-1?

```python
N = 4
ts = numpy.arange(N) / N
freqs = numpy.arange(0, N)
args = numpy.outer(ts, freqs)
M1 = numpy.exp(1j * PI2 * args)
print(M1)
```

```python
N = 4
ts = numpy.arange(N) / N
freqs = numpy.arange(N, 2*N)
args = numpy.outer(ts, freqs)
M2 = numpy.exp(1j * PI2 * args)
print(M2)
```

```python
def approx_equal(a, b, tol=1e-10):
    return sum(abs(a-b)) < tol
```

```python
approx_equal(M1, M2)
```

This is true! The frequencies going from 0 to N-1 equals the matrix representing the frequencies from N to 2*N -1. This is because the signal is periodic. At every snapshot of the period, the amplitudes will be approximately equal. This is confirmed by the output of the approx_equal function. When plotting the amplitudes, I noticed something tricky. The output amplitudes were the same as the input amplitude, but in the wrong order. I thought this was interesting and worth looking into. 

#### 2) In thinkdsp.py, the Spectrum methods low_pass and high_pass are ideal filters, in some sense: in the passband they are perfectly flat, and in the stopband they attenuate to 0 (or whatever value you want).  Invoking the principle that for every filter there is a corresponding window, see if you can figure out what window corresponds to these ideal filters.

```python
sawtooth = thinkdsp.SawtoothSignal(freq = 440, amp =1)
sawtooth.plot()
sawtooth_amps = sawtooth.make_wave().make_spectrum().amps
```

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res
```

```python
def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(1j * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys
```

```python
a = numpy.ones(50)
a = zero_pad(a, 100)
ts_a = numpy.linspace(0, 25, 100)
freqs = numpy.linspace(0, 1, 100)
ys_a = synthesize2(a, freqs,ts_a)
```

```python
b = []
b1 = zero_pad(b, 50)
b2 = numpy.ones(50)
b = numpy.concatenate((b1, b2))
ts_b = numpy.linspace(0, 25, 100)
freqs = numpy.linspace(0, 1, 100)
ys_b = synthesize2(b, freqs,ts_b)
```

```python
thinkplot.preplot(2)
print(len(ts_a))
print(len(ys_a))
thinkplot.plot(ts_a, ys_a, label = r'ihwihw')
thinkplot.plot(ts_b, ys_b, label=r'high pass window')
```

This was a fun problem! Essentially, in the frequency domain, we can make an ideal filter that is high for however long we want and then low for also however long we want. We can make this filter in the frequency domain. Then we can take the inverse dft function to make the filter into a window in the time domain. This window looks like a $\frac{sin(x)}{x}$ function or a sinc function. We can tell that the amplitudes are getting smaller with time, this means this does sinc. We can see that low pass filter is a periodic sinusoid, while the high pass filter has more spikes and it looks more complex. This is probably because the high pass filter has more energy in the same time snippet, in comparison to the low pass filter. 

#### 3) So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.

```python
signal = thinkdsp.SquareSignal(freq=440)
wave = signal.make_wave(duration=1, framerate=44100)
wave_seg = wave.segment(start = 0, duration = 0.01)
```

```python
#plot phase offset as a function frequency before convolving 
N = len(wave_seg.ys)
fs_1 = numpy.arange(N) * wave_seg.framerate / N

smooth_hs_no_convolve = numpy.fft.fft(wave_seg.ys)
smooth_hs = numpy.fft.fft(wave_seg.ys)
phase = numpy.angle(smooth_hs)
thinkplot.plot(fs_1, phase)
```

```python
##convolved wave
#gaussian window
gaussian = scipy.signal.gaussian(M=11, std=0.8)
gaussian /= sum(gaussian)

N = len(wave_seg.ys)
fs_2 = numpy.arange(N)*wave_seg.framerate/N

##take fft 
ys_gaussian = numpy.convolve(wave_seg.ys, gaussian, mode='same')
smooth_gaussian_hs = numpy.fft.fft(ys_gaussian)
phase_gaussian_2 = numpy.angle(smooth_gaussian_hs)

thinkplot.plot(fs_2, phase_gaussian_2)
```

```python
##convolved wave
#gaussian window
boxcar = scipy.signal.boxcar(M=10)
boxcar /= sum(boxcar)

N = len(wave_seg.ys)
fs_2 = numpy.arange(N)*wave_seg.framerate/N

##take fft 
ys_boxcar = numpy.convolve(wave_seg.ys, boxcar, mode='same')
smooth_boxcar_hs = numpy.fft.fft(ys_boxcar)
phase_boxcar_2 = numpy.angle(smooth_boxcar_hs)

thinkplot.plot(fs_2, phase_boxcar_2)
```

There is no difference! They look the same! The phase shift as a function of the frequency of a normal wave is the same as the phase shift as a function of the frequency of the wave convolved with the gaussian wave and the phase shift as a function of the frequency of the wave convolved with the boxcar wave. 

#### 4) Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this window is an increasing function of f, so you can think of diff as a high-pass filter.  One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.  If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.  Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?  See also http://en.wikipedia.org/wiki/Finite_difference_coefficient

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def window_to_filter(window): 
    array = window
    padded_array = zero_pad(array, 500)
    new_wave = thinkdsp.Wave(padded_array, framerate = 10000)
    new_wave.make_spectrum().plot()
```

```python
window_to_filter([-1, 1]) ##aproximation of the differentiation of the signal 
```

```python
window_to_filter([-1/2, 2, -3/2])
```

```python
window_to_filter([1, -2, 1])
```

The first window filter is the approximation of the difference of the signal. This is approximately a high pass filter.The second window filter is a better approximation of the difference because we have a tighter fit. The last window is the derivative of the signal, which is approximately a high pass filter as well. We can see that the last window, the [1, -2, 1] is the best derivative approximation.
