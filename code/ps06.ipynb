```python
import numpy
import math 
import thinkplot
import numpy
import matplotlib.pyplot as plt
```

```python
def sinc_function(ts, T, K):
    x_sub_k = np.zeros(len(ts))
    ks = np.arange(-K/2+1, K/2 +1)
    for n in range(len(ks)):
        x_sub_k +=numpy.sinc(n/2.0)*numpy.exp(1j*(2.0*numpy.pi/T)*n*ts)*0.5
    return x_sub_k
```

```python
K1 = 5
K2 = 17 
K3 = 257
ts = numpy.linspace(-6, 6, 100)
ys1 = sinc_function(ts, 4, K1)
ys2 = sinc_function(ts, 4, K2)
ys3 = sinc_function(ts, 4, K3)
plt.subplot(3, 1, 1)
plt.hold(True)
plt.plot(ts, ys1)
plt.title('Fourier Series for K = 5 terms')
plt.subplot(3, 1, 2)
plt.hold(True)
plt.plot(ts, ys2)
plt.title('Fourier Series for K = 17 terms')
plt.subplot(3, 1, 3)
plt.hold(True)
plt.plot(ts, ys3)
plt.title('Fourier Series for K = 257 terms')
```
