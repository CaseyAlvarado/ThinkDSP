If we simply add a phase offset to the complex exponential, in the frequency domain, the phase offset will make the wave start at different points and therefore, we see the second wave (the wave with the offset) rotate around the first wave. However, if we add the phase offset to the time constant, then we will see the wave with the offset shift in the time domain. We can show this with math below.

## 2) In Chapter 7 I represented the DFT as matrix multiplication, but the more common presentation is as a summation:  From http://en.wikipedia.org/wiki/Discrete_Fourier_transform:   Read the Wikipedia page on DFT.  Write an implementation of DFT based on this summation and compare your results to my implementation and/or numpy.fft.fft.

```python
def dft_imp(ys): 
    N = len(ys)
    freq = numpy.arange(N)
    x_sub_n = ys
    su = 0 
    for n in range(0, N-1): 
        s = x_sub_n[n]*numpy.exp((-1j*PI2*n*freq)/N)
        su = su+s
    return su
```

## 3) In Chapter 7, I showed how we can express the DFT and inverse DFT as matrix multiplications. These operations are relatively fast, taking time proportional to N2, where N is the length of the wave array. That would be fast enough for many applications, but it turns out that there is a faster algorithm, the Fast Fourier Transform (FFT), which takes time proportional to N logN.  Read about the FFT at http://en.wikipedia.org/wiki/Cooley-Tukey_FFT_algorithm, and write an implementation. Hint: I suggest you write a simple version as a recursive function; don’t worry about “data reordering, bit reversal, and in-place algorithms”.

```python
def fft(ys):  
    N = len(ys)
    if N ==1: 
        return ys
    else:
        even = fft([ys[i] for i in range(0, N, 2)])
        odd = fft([ys[i] for i in range(1, N, 2)])
        holder = [0] * N
        
        for m in range(int(N/2)): 
            holder[m] = even[m] + numpy.exp((PI2*-1j*m)/N) * odd[m]
            holder[m+int(N/2)] = even[m] - numpy.exp((PI2*-1j*m)/N) * odd[m]
        return holder 
```

## 4) Using the timing code from PS03, compare the performance of my implementation of autocorr from Chapter 5, my fft_autocorr from Chapter 8, and numpy.correlate.  Can you tell whether NumPy is using the FFT algorithm?

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)

    corrs = fft_convolve(signal, window)
    corrs = corrs[N//2: 3*N//2]
    return corrs
```

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def fft_convolve(signal, window):
    fft_signal = numpy.fft.fft(signal)
    fft_window = numpy.fft.fft(window)
    return numpy.fft.ifft(fft_signal * fft_window)
```

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=2.0)

ns = 2 ** numpy.arange(4, 12)
print(ns)
```

```python
results_autocorr = []
results_fft_autocorr = []
results_numpy_correlate = []

for N in ns:
    signal = thinkdsp.UncorrelatedGaussianNoise()
    noise = signal.make_wave(duration=1.0)
    noise.ys = noise.ys[:N]
    new_wave = thinkdsp.Wave(noise.ys, noise.framerate)
    result_autocorr = %timeit -o autocorr(noise)
    result_fft_autocorr = %timeit -o fft_autocorr(noise.ys[:N])
    results_autocorr.append(result_autocorr)
    results_fft_autocorr.append(result_fft_autocorr)
    result_numpy_correlate = %timeit -o numpy.correlate(noise.ys[:N], noise.ys[:N], 'full')
    results_numpy_correlate.append(result_numpy_correlate)
    
    
bests_autocorr = [result_autocorr.best for result_autocorr in results_autocorr]
bests_fft_autocorr = [result_fft_autocorr.best for result_fft_autocorr in results_fft_autocorr]
bests_numpy_correlate = [result_numpy_correlate.best for result_numpy_correlate in results_numpy_correlate]
bests_autocorr
bests_fft_autocorr
#bests_numpy_correlate
```
