2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

lags,corrs = autocorr(wave)
thinkplot.plot(lags, corrs)
thinkplot.config(xlabel = 'lags', ylabel = 'corrs')
```

This data does not decay exponentially. The first region looks like it is decaying linearly from the starting point. I noticed a lag of 0 and a correlation of 1. This makes sense because any wave right on top of itself has a high correlation because it is itself. There is a spike at a lag value of around 500Hz, this means there is evidence of periodic behavior. 
This period of the wave is lag/framerate and in this case, the framerate is 1 segment/second and the lag peak is at 500 segments. 


Period = $\frac{lag}{framerate}$ = $\frac{500}{1}$


Frequency = $\frac{1}{Period}$ = $\frac{1}{500}$

3) Use autocorrelation to estimate the pitch of the saxophone recording in 100475__iluppai__saxophone-weep.wav.  What happens if you filter out the fundamental tone?  Is the estimate based on the autocorrelation function consistent with your perceived pitch?

```python
sax_full = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
sax_seg = sax_full.segment(start = 0.5, duration = 1)
sax_seg.make_audio()
```

```python
lags, corrs = autocorr(sax_seg)
thinkplot.plot(lags, corrs)
thinkplot.config(xlim = [0, 1000], ylim = [-0.5, 1])
```

There are multiple peaks, which means this signal is periodic and it also looks like this signal has both even and odd harmonics. The first peak is at around 120 segment in lag. We can calculate the period and then the frequency from this data. 


Period = $\frac{lag}{framerate}$ = $\frac{120 segments}{11025 segments per second}$

Frequency = $\frac{1}{Period}$ = $\frac{1}{\frac{120}{11025}}$ = \frac{11025}{120} = 91.87 Hz

5) One of the major applications of the DCT is compression for both sound and images. In its simplest form, DCT-based compression works like this:

1. Break a long signal into segments.
2. Compute the DCT of each segment.
3. Identify frequency components with amplitudes so low they are inaudible, and remove them. Store only the frequencies and amplitudes that remain.
4. To play back the signal, load the frequencies and amplitudes for each segment and apply the inverse DCT.

Implement a version of this algorithm and apply it to a recording of music or speech. How many components can you eliminate before the difference is perceptible?

```python
def notAudible(element):
    return element > 1500

def inverse_dct_iv(amps): 
    return dct.dct_iv(amps)

def compression(w): 
    ys = []
    #wave = w.scale(0.1)
    woops = thinkdsp.SinSignal(freq = 1, amp = 0)
    wave = woops.make_wave(framerate = 44100)
    for i in (numpy.linspace(0, w.duration, w.duration/0.01)):
        seg = w.segment(start = i, duration = 1/0.01)
        framerate = seg.framerate
        d = dct.dct_iv(seg.ys) 
        audible_amps = filter(notAudible, d)
        ys.extend(inverse_dct_iv(audible_amps))
    wave = thinkdsp.Wave(ys, framerate = framerate)
    return wave
```

```python
noise = thinkdsp.read_wave('noise.wav')
noise_segment = noise.segment(start = 0.02, duration = 0.4)
noise_segment.plot()
noise_segment.make_audio()
```

```python
compressed_wave = compression(noise_segment)
compressed_wave.plot()
compressed_wave.make_audio()
```
