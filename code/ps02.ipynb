2) Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly. Hint: combine the evaluate functions from Chirp and SawtoothSignal.
Plot the spectrogram of this signal. The effect of aliasing should be visible, and if you listen carefully, you can hear it.

```python
class SawtoothChirp(thinkdsp.Chirp): 
    def evaluate(self, ts):
        freqs = numpy.linspace(self.start, self.end, len(ts))
        A = 1/freqs 
        ys  = 2*((ts/A) - numpy.floor(0.5 + (ts/A)))
        return ys
```

```python
signal = SawtoothChirp(start=220, end=880)
wave = signal.make_wave(duration=2, framerate=10000)
wave.apodize()
wave.make_audio()
```

```python
sp = wave.make_spectrogram(1024)
sp.plot()
thinkplot.config(xlabel='time (s)', ylabel='frequency (Hz)', legend=False)
```

3)  In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length.

Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp?

Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.

```python
class G(thinkdsp.Chirp): 
    def glissando(self, ts): 
        frequency = numpy.linspace(self.start, self.end, len(ts))
        return self._evaluate(ts, frequency).make_wave()
```

```python
def playGlissando(low, high):
    signal1 = G(low, high)
    wave1 = signal1.make_wave()

    signal2 = G(high, low)
    wave2 = signal2.make_wave()

    new_wave = wave1 | wave2
    return new_wave
```

```python
out_wave = playGlissando(262, 349)
out_wave.apodize()
out_wave.make_audio()
```

```python
out_wave.make_spectrogram(1024).plot(high = 50)
```

5) “A Soft Murmur” is a web site that plays a mixture of natural noise sources, including rain, waves, wind, etc. At http://asoftmurmur.com/about/ you can find their list of recordings, most of which are at http://freesound.org.

Download a few of these files and compute the spectrum of each signal. Does the power spectrum look like white noise, pink noise, or Brownian noise? How does the spectrum vary over time?

```python
signal = thinkdsp.read_wave('noise.wav'); 
seg = signal.segment(start = 0, duration = 15)
seg.make_audio()
```

```python
spectrumm = seg.make_spectrum()
spectrumm.plot()
thinkplot.show(xscale = 'log', yscale = 'log')
spectrumm.estimate_slope()
```

The spectrogram has different characteristics in different regions. In the mili-seconds range, the frequency remains stable in the MHz region. In the 10^0 to 10^2 second range, the frequency also remains stable, with more noise, at 10^5 Hz. Finally, in the 10^2 to 10^4 second range, the signal decreases. 

According to the estimate_slope() function, it decreases at the rate of -2.6335145230996422. 

Because this graph is on a log scale, we can develop a relationship between the power and the frequency in the $y = m*x + b$ format. 

Therefore, $log(Power) = m*log(Frequency) - b$, or $log(power) = -2.6335145230996422*log(frequency).$

We can use log rules, $log(power) = log(\frac{b}{frequency^-2.6335145230996422})$ and we can get rid of the log on both sides and acquire: power = $\frac{b}{frequency^(-2.6335145230996422)}$. 

We can see that the power is inversely proportional to the frequency squared. Beta is $-2.6335145230996422$. Brownian noise beta is around $2$. This means that this is Brownian noise, or very close to it!

6) In a white noise signal, the mixture of frequencies changes over time. In the long run, we expect the power at all frequencies to be equal, but in any sample, the power at each frequency is random.

To estimate the long-term average power at each frequency, we can break a long signal into segments, compute the power spectral density for each segment, and then compute the average across the segments. 

You can read more about this algorithm at http://en.wikipedia.org/wiki/Bartlett's_method.

Apply Bartlett's method to your natural noise source and see if it gives a clearer picture of the relationship between frequency and power.


```python
def BartlettsMethod(w):
    summs = 0
    freq = 0
    for i in (numpy.linspace(0, w.duration - 1, w.duration)):
        seg = w.segment(start = i, duration = 1)
        p = seg.make_spectrum().power
        summs = summs + p
        freq = freq + seg.make_spectrum().fs
    average = summs/(w.duration - 1)
    return average, freq
```

```python
a , f = BartlettsMethod(seg)
thinkplot.plot(f, a)
thinkplot.config(xlim = [0, 15000], xlabel = 'frequency in Hz', ylabel = 'power (amps^2')
```
